```python
x = np.array([10,20,30,40,50,60])
print(x.ndim)
print(x.shape)
print("x의값 : ",x)
```

```python
x = np.array( [  [ 11, 21, 31 ],
                 [ 12, 22, 32 ],
                 [ 13, 23, 33 ]   ]  )
print(x.ndim)
print(x.shape)
print("x의값 :",x)
```

```python
import mglearn

import numpy as np
import matplotlib.pyplot as plt

line = np.linspace(-3, 3, 100)
tanh_line = np.tanh(line)
relu_line = np.maximum(line, 0)  # 두개의 배열값 중 최대값 찾기
sig_line = 1/(1+np.exp(-line))

step_line = line.copy()
step_line[step_line <= 0] = 0
step_line[step_line > 0] = 1
```

```python
# 음수 표시
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.rcParams["figure.figsize"] = (14,10)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
# step function
ax1.plot(line, step_line, label='step', color='red')
ax1.legend(loc='best')

# sigmoid function (시그모이드 함수)
ax2.plot(line, sig_line, label='sigmoid', color='green')
ax2.legend(loc='best')

# Hyperbolic Tangent(tanh)
ax3.plot(line, tanh_line, label='tanh')
ax3.legend(loc='best')

# relu
ax4.plot(line, np.maximum(line, 0), label='relu', color='orange')
ax4.legend(loc='best')
```

## 기본 딥러닝 모델 구축하기
* 입력과 출력으로 데이터를 나누어 주어야한다.
* 통상적으로 이미지에 대해서 다루게 될때, 입력은 이미지 정보가 된다.

  * 입력 - 이미지 정보
  * 출력 - 예측하고자 하는 어떠한 값.

```python
train.columns, test.columns
```

```python
sel = ['weather', 'temp','season','windspeed','workingday', 'hour', 'dayofweek']
X = train[sel]
y = train['count']

X_val = test[sel]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(32, input_dim = 7 ,activation = 'relu' ))
model.add(Dense(16, activation = 'relu'))
model.add(Dense(16, activation = 'relu'))
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(loss = 'mean_squared_error', optimizer = 'rmsprop')
model.fit(X_train, y_train,
          validation_data=[X_test, y_test],
          epochs=100, batch_size=10)
```

```python
model.evaluate(X_test, y_test)
```

```python
pred = model.predict(X_val)
pred [ 0: 100]
```

```python
sub = pd.read_csv("sampleSubmission.csv")
sub['count']=pred

sub.loc[sub['count']<0, 'count'] = 0
```
